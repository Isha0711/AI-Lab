```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns 
```

```python
df= pd.read_csv("/Users/ishhh/Downloads/anime.csv")
df.head(10)
```

```python
df.dtypes
```

```python
#Dropping columns with a lot of null values(57% japanese name and 85% end year missing in the dataset)
df= df.drop(['Japanese_name', 'End_year'],axis=1)
```

```python
df.head(10)
```

```python
#boxplot
sns.boxplot(x=df['Episodes'])
```

```python
sns.boxplot(y=df['Episodes'])
```

```python
df.describe()
```

```python
#scatter plot
df.plot.scatter(x='Rank', y='Rating',title='Scatter Plot: Rank vs Rating', alpha=0.5)
plt.show()
```

```python
#scatter plot with release year as well
scatter_plot = plt.scatter(
    x=df['Rank'], y=df['Rating'], c=df['Release_year'], cmap='viridis', s=100)

cbar = plt.colorbar(scatter_plot, label='Release_year')

plt.xlabel('Rank')
plt.ylabel('Rating')
plt.title('Scatter Plot: Rank vs Rating (colored by Release_year)')

plt.show()
```

```python
#barplot for the top 10 data
sns.barplot(x='Episodes', y='Rank', hue='Type', data=df.head(10))
```

```python
#displot of rank and rating
sns.distplot(df[df['Rating']>4]['Rank'],hist=True)
sns.distplot(df[df['Rating']<=4]['Rank'],hist=True)
```

```python
# horizontal bar plot
df_filtered = df[df['Release_year'] > 2010]
df_filtered['Release_year'].value_counts().sort_index().plot(
    kind='barh', figsize=(10, 8), color='skyblue', edgecolor='black')

plt.title("Number of Animes Released Each Year since 2010")
plt.ylabel("Year")
plt.xlabel("Number of Animes")

plt.show()
```

```python
#Heat Map

# Filter data for release years since 2010 and top 10 ranks
df_filtered = df[(df['Release_year'] > 2010) & (df['Rank'] <= 10)]

# Create a crosstab of Release_year and Rank
crosstab_data = pd.crosstab(df_filtered['Release_year'], df_filtered['Rank'])

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(crosstab_data, cmap='Blues', annot=True,
            fmt='g', cbar_kws={'label': 'Number of Animes'})
plt.title('Heatmap of Top 10 Animes Released Each Year and Rank (since 2010)')
plt.xlabel('Rank')
plt.ylabel('Year')

plt.show()
```
